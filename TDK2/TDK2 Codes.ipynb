```r
library(R2DAG)
library(ConnectednessApproach)
library(relaimpo)
library(pcalg)
```

## Example 1

```r
A0 <- matrix(c(1, 0, 0,
                0, 1, 0,  
                -2, -2, 1), nrow = 3, byrow = TRUE)

data <- gen_svar(1000, A0=A0, list_A=list(), df=20)
```

```r
DR2 <- R2_network(data, "genizi", directed=T)
DR2
```

```r
R2 <- R2_network(data, "genizi", directed=F)
R2
```

```r
plot_irf_graph(DR2$table, name ="DAG for DR^{2G}", vertex_label_size = 2)
```

```r
plot_irf_graph(R2$table, name ="DAG for R^{2G}", vertex_label_size = 2)
```

```r
r2 <- readRDS("r2_tci.rds")
dr2 <- readRDS("dr2_tci.rds")
```

```r
#png("empirical.png", width = 1200, height = 600)

# Set larger margins: bottom, left, top, right
par(mar = c(5, 6, 4, 0))  # default is c(5, 4, 4, 2)

# Extract date index
date_index <- as.Date(rownames(as.data.frame(dy2012))[253:2771])

# Plot
plot(
  x = date_index,
  y = r2$tci_vector,
  type = "l",
  col = "red",
  lwd = 2,
  ylim = c(0, 1),
  xlab = "Date",
  ylab = "TCI",
  main = "",
  cex.axis = 2,
  cex.lab = 2,
  cex.main = 2
)

lines(date_index, dr2$tci_vector, col = "blue", lwd = 2)

legend("topright",
       legend = c(expression(R^{2*G}), expression(D*R^{2*G})),
       col = c("red", "blue"),
       lwd = 2,
       cex = 2)

#dev.off()
```
